````python
x1      = dem_ran100.demographics_mean_reation_time_ms
x2      = dem_top100.demographics_mean_reation_time_ms
cols    = ['#2f547d', '#8B2252']
cols    = ['steelblue', 'orange']


def plot_demographics(x1,x2, x_label='Reaction times (RT)', y_label='Frequency', cols=['#E2725A', 'steelblue'], nmes=['Random', 'Top']):
    fig = go.Figure()
    for x, c, n in zip([x1,x2], cols, nmes):
        fig.add_trace(go.Histogram(x=x, histnorm='percent', marker_color=c, name=n, opacity=0.7))
    fig.update_layout(template='none', xaxis_title=x_label, yaxis_title=y_label, legend_title="Sample", font=dict(size=15), width=850, height=500)
    fig.show()


# D294AF, 79AEB2, E2725A
plot_demographics(x1,x2)
````

````python
# Income distribution ---------------------------------*
x1 = dem_top100.demographics_income_a
x2 = dem_ran100.demographics_income_a
plot_demographics(x1,x2,x_label='Income')

# CS hours distribution -------------------------------*
x1 = dem_top100.demographics_computer_hours_a
x2 = dem_ran100.demographics_computer_hours_a
plot_demographics(x1,x2,x_label='Computer hours')


# Income distribution ---------------------------------*
x1 = dem_top100.demographics_age_a
x2 = dem_ran100.demographics_age_a
plot_demographics(x1,x2,x_label='Age')
````
