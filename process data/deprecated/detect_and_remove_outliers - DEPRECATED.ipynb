# Outlier detection & removal

A subset of subjects may have completed the experiment however, in haste or negligance, failed to adequentely complete each task.

This can be readily observed in the data given the simple nature of the tasks. We are not studying participants with mental defects and as such extremely poor performance can be considered gross neglicance, task incompletion or corrupt data.


## Thresholding

A performance threshold should be chosen for each task, achieving results under this minimum should exclude the subjects data for the given experiment.


**This is such a small addition it is best added during the modeling process.**
